# Import Libraries and Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


df = sns.load_dataset('anscombe')
df
```

# Creating the Anscombeâ€™s quartet - Object Oriented Programming

```python
# Creating 4 separate subplots or axes for each quartet to be plotted and reside in 
fig, ax =  plt.subplots(2,2, figsize = [15,8])

"--------------------------------Quartet 1-------------------------------------------"
# Defining the dataset based the 1st quartet (Quartet I) that is indicated within the "dataset" column
df1 = df[df['dataset'] == 'I']
# Defining the two columns of data that will be used for analysis
x = df1['x']
y = df1['y']


# Mean and standard deviation (both streams of data, x and y) for each quartet
# is calculated using numpy and will be shown in text form on the plots
mean_df1_x = np.mean(x)
mean_df1_y = np.mean(y)
stddev_df1_x = np.std(x)
stddev_df1_y = np.std(y)

# Set the same x and y limits for each quartet within each subplot
ax[0,0].set_xlim(0,20)
ax[0,0].set_ylim(0,14)

# This straight_line function creates the best line of fit that will be used 
# and plotted within each subplot, separately for each quartet
def straight_line(x,y):
    a, b  = np.polyfit(x,y,deg=1)
    line  = a * x + b
    return line
line = straight_line(x,y)

# The particular quartet data is plotted withinthe axes specified with indication
# of marker, color and edgecolor of data points as well as the plotting of the best line of fit
ax[0,0].plot(x,y, 
            linestyle = '', marker = 'o',
            color = 'orange', markersize = 12,
            markeredgecolor = 'red') 
ax[0,0].plot(x, line, 'b-')

# This is the labeling of the particular subplot including the x and y-axis labels, title, and text
# for the mean and standard deviations for x and y columns of data that is specified with each subplot
ax[0,0].text(0,1, 'Mean for X = %s\n Mean for Y = %s\n Std Dev for X = %s\n Std Dev for Y = %s' % (mean_df1_x, 
            mean_df1_y, stddev_df1_x, stddev_df1_y))
ax[0,0].set_xlabel('$x_{1}$')
ax[0,0].set_ylabel('$y_{1}$')
ax[0,0].set_title("Anscombe's quartet - I")

# Be mindful that the following blocks of code for each quartet repeats the same process
"----------------------------------------------Quarter 2-------------------------------------------------------"

df2 = df[df['dataset'] == 'II']
x2 = df2['x']
y2 = df2['y']

mean_df2_x = np.mean(x2)
mean_df2_y = np.mean(y2)
stddev_df2_x = np.std(x2)
stddev_df2_y = np.std(y2)

ax[0,1].set_xlim(0,20)
ax[0,1].set_ylim(0,14)
                     
def straight_line(x2,y2):
    a, b  = np.polyfit(x2,y2,deg=1)
    line  = a * x2 + b
    return line
line = straight_line(x2,y2)

ax[0,1].plot(x2,y2, 
            linestyle = '', marker = 'o',
            color = 'purple', markersize = 12,
            markeredgecolor = 'blue') 
ax[0,1].plot(x2, line, 'b-')

ax[0,1].text(0,1, 'Mean for X = %s\n Mean for Y = %s\n Std Dev for X = %s\n Std Dev for Y = %s' % (mean_df2_x, 
            mean_df2_y, stddev_df2_x, stddev_df2_y))
ax[0,1].set_xlabel('$x_{2}$')
ax[0,1].set_ylabel('$y_{2}$')
ax[0,1].set_title("Anscombe's quartet - II")


"------------------------------------------------------Quartet 3--------------------------------------------------------"
df3 = df[df['dataset'] == 'III']
x3 = df3['x']
y3 = df3['y']

mean_df3_x = np.mean(x3)
mean_df3_y = np.mean(y3)
stddev_df3_x = np.std(x3)
stddev_df3_y = np.std(y3)

ax[1,0].set_xlim(0,20)
ax[1,0].set_ylim(0,14)
                     
def straight_line(x3,y3):
    a, b  = np.polyfit(x3,y3,deg=1)
    line  = a * x3 + b
    return line
line = straight_line(x3,y3)

ax[1,0].plot(x3,y3, 
            linestyle = '', marker = 'o',
            color = 'red', markersize = 12,
            markeredgecolor = 'purple') 
ax[1,0].plot(x3, line, 'b-')

ax[1,0].text(0,1, 'Mean for X = %s\n Mean for Y = %s\n Std Dev for X = %s\n Std Dev for Y = %s' % (mean_df3_x, 
            mean_df3_y, stddev_df3_x, stddev_df3_y))
ax[1,0].set_xlabel('$x_{3}$')
ax[1,0].set_ylabel('$y_{3}$')
ax[1,0].set_title("Anscombe's quartet - III")
                     
"------------------------------------------------------Quartet 4--------------------------------------------------------"
df4 = df[df['dataset'] == 'IV']
x4 = df4['x']
y4 = df4['y']

mean_df4_x = np.mean(x4)
mean_df4_y = np.mean(y4)
stddev_df4_x = np.std(x4)
stddev_df4_y = np.std(y4)

ax[1,1].set_xlim(0,20)
ax[1,1].set_ylim(0,14)
                     
def straight_line(x4,y4):
    a, b  = np.polyfit(x4,y4,deg=1)
    line  = a * x4 + b
    return line
line = straight_line(x4,y4)

ax[1,1].plot(x4,y4, 
            linestyle = '', marker = 'o',
            color = 'green', markersize = 12,
            markeredgecolor = 'yellow') 
ax[1,1].plot(x4, line, 'b-')

ax[1,1].text(0,1, 'Mean for X = %s\n Mean for Y = %s\n Std Dev for X = %s\n Std Dev for Y = %s' % (mean_df4_x, 
            mean_df4_y, stddev_df4_x, stddev_df4_y))
ax[1,1].set_xlabel('$x_{4}$')
ax[1,1].set_ylabel('$y_{4}$')
ax[1,1].set_title("Anscombe's quartet - IV")
'--------------------------------------------------------------------------------------------------------------------------'
# Organizes the subplots in a very neat and compact matter in which text does not overlap with each other
plt.tight_layout()
```
