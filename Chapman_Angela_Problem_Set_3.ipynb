# Web Scraping Data

```python
url = 'https://en.wikipedia.org/wiki/MonsterVerse'
df = pd.read_html(url)

# Selecting the particular table (#4)vas specified in the Problem Set Instructions
df = df[3]
df
```

# Cleaning Up the DataFrame Steps 1-3 Function

```python
def clean_table(df):
    # Turn MultiIndex columns, or tuples, into a list of tuples
    df_list = list(df.columns)
    # Empty movie list act as a way for all the movie column headers to be renamed to proper single column header
    movies = []
    # This index allows access to the 2nd level of the column headers, aka gives access to the movies
    ind = 1

    for col in df_list:
        # Access to the particular movie out of the column headers
        if col[ind] == 'Character':
            # Setting the charcter column as the index and dropping the original column 
            df.index = df[df.columns[0]]
            df.index.name = 'character'
            df = df.drop(columns = ['Character'])
        else:
            # Creating movies list so that the new columns can be created later
            movies.append(col[ind])
    
    # Setting the new column as single column headers for the movies instead of the MultiIndex
    df.columns = movies
    # df.drop("Humans") could not do, could not figure out how to drop the humans row and apply to fast & furious
                          

            
    return df
```

# Running Function 

```python
clean_table(df)
```

# Saving as CSV File (Step 4)

```python
df.to_csv('characters.csv', sep = '\t')
```
